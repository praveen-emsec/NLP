```python
import spacy
from spacy import displacy
from spacy.lang.en import English
```

```python
from spacy.pipeline import EntityRuler
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
import pandas as pd
```

```python
df = pd.read_csv('20_newsgroups_cleaned_data_1.csv')
```

```python
docs = []
for index, row in df.iterrows():
    text = row['data']
    doc = nlp(text)
    docs.append(doc)
```

```python
for doc in docs:
  for ent in doc.ents:
    print(ent.text,ent.label_,ent.vector_norm)
```
