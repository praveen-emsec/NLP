```python
import pandas as pd
df1=pd.read_csv('times.csv')
```

```python
from simpletransformers.language_representation import RepresentationModel
        
input_data = df1["Title"].tolist()

model = RepresentationModel(
        model_type="bert",
        model_name="bert-base-uncased",
        use_cuda=False
    )

sentence_vectors = model.encode_sentences(input_data, combine_strategy='mean')
```

```python
for sentence,vector in zip(input_data,sentence_vectors):
  print(f"sentence:{sentence}+\n+Vectors:{vector}")
```
