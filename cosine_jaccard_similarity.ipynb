```python
# Read the CSV file and store the data in an array or list
df = pd.read_csv('medium_Data.csv')
```

```python
df = df.dropna(subset=["Contents"])
```

```python
print(len(df['Titles']))
print(len(df['Contents']))
```

```python
vectorizer = TfidfVectorizer()
vector1 = vectorizer.fit_transform(df["Titles"])
vector2 = vectorizer.fit_transform(df["Contents"])
vector2 = np.resize(vector2, (298, 1110))
print(vector1.shape)
print(vector2.shape)
```

```python
# Compute the cosine similarity between the data arrays
start_time = time.time()
cosine_sim = cosine_similarity(vector1)
cosine_time = time.time() - start_time

# Compute the Jaccard similarity between the data arrays
```

```python
reader = df.to_numpy().tolist()

print(reader)
reader = [set(ele) for ele in reader]
     
```

```python
def jaccard_similarity(set1, set2):
    return len(set1.intersection(set2)) / len(set1.union(set2))
```

```python
jaccard_similarity_scores = []
start_time = time.time()
for i in range(len(reader)):
    for j in range(i + 1, len(reader)):
        set1 = reader[i]
        set2 = reader[j]
        similarity_score = jaccard_similarity(set1, set2)
        jaccard_similarity_scores.append(similarity_score)
jaccard_time = time.time() - start_time
print(jaccard_similarity_scores)
```
