```python
import pandas as pd
df1= pd.read_csv('times.csv')
```

```python
# Import the Pandas and Gensim libraries
import pandas as pd
import gensim


text = df1["Summary"]
words = [x for x in text.str.split()]
model = gensim.models.Word2Vec(words, size=100, window=5, min_count=1, workers=4)

word_vectors = [model[word] for word in words]
```

```python
for word,vectors in zip(words,word_vectors):
  print(f"Word:{word}, Vectors:{vectors}")
```
