```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
import pandas as pd
df=pd.read_csv("medium_Data.csv")
df
```

```python
import re
import string

alphanumeric = lambda x: re.sub(r"""\S*\d{1}\w*""", ' ', x)
punc_lower = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x.lower())
new_line = lambda x: re.sub(r'(?<!\.)\n', '', x)
df=df.astype(str)
df['Titles'] = df.Titles.map(alphanumeric).map(punc_lower).map(new_line)
df['Contents'] = df.Contents.map(alphanumeric).map(punc_lower).map(new_line)
df.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english'))
```

```python
df['Titles'] = df['Titles'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df['Contents'] = df['Contents'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
df.to_csv("df1.csv", index=False)
```

```python
# # Create a new dataframe with the named entities as separate columns
# ner_df = pd.DataFrame(columns=['PERSON', 'ORG', 'GPE','DATE','NORP'])
```

```python
entities = []
for index, row in df.iterrows():
    title = row['Titles']
    content = row['Contents']
    
    doc = nlp(title + ' ' + content)
    
    # Extract named entities
    entities.extend([(entity.text, entity.label_) for entity in doc.ents])
entities  
```

```python
ner_dict = {}
```

```python
for entity, label in entities:
    # If the label is not in the dictionary, create a new entry
    if label not in ner_dict:
        ner_dict[label] = []
    # Append the named entity to the list for the corresponding label
    ner_dict[label].append(entity)
```

```python
ner_df = pd.DataFrame.from_dict(ner_dict, orient='index')
ner_df = ner_df.transpose()
```
