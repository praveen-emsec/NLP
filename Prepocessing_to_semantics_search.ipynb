```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
def removestopword(text): 
  sentence = nlp(text)
  notStopWords = [notStopWords.text for notStopWords in sentence if not notStopWords.is_stop]  
  joined = " ".join(notStopWords)
  return joined
```

```python
df=df.astype(str)
df=df.applymap(removestopword)
df.head()
```

```python
entities = []
for index, row in df.iterrows():
    heading = row['Heading']
    description = row['Description']
    
    doc = nlp(heading + ' ' + description)
    
    # Extract named entities
    entities.extend([(entity.text, entity.label_) for entity in doc.ents])
entities 
```

```python
ner_dict = {}
for entity, label in entities:
    # If the label is not in the dictionary, create a new entry
    if label not in ner_dict:
        ner_dict[label] = []
    # Append the named entity to the list for the corresponding label
    ner_dict[label].append(entity)
```

```python
ner_df = pd.DataFrame.from_dict(ner_dict, orient='index')
ner_df = ner_df.transpose()
ner_df.head(20)
```

```python
def concatenate_text(examples):
    return {
        "text": examples["Heading"]
        + " \n "
        + examples["Description"]
        + " \n "
        
    }

dataset = dataset.map(concatenate_text)
```

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer, TFAutoModel

model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = TFAutoModel.from_pretrained(model_ckpt, from_pt=True)
```

```python
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
```

```python
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, return_tensors="tf"
    )
    encoded_input = {k: v for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output)
```

```python
embedding = get_embeddings(dataset["text"][0])
embedding.shape
```

```python
embeddings_dataset = dataset.map(
    lambda x: {"embeddings": get_embeddings(x["text"]).numpy()[0]}
)
```
